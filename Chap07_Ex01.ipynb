```python
#ライブラリのインポート
import numpy as np
import numpy.random as random
import scipy as sp
import pandas as pd
from pandas import Series, DataFrame
import requests
import zipfile
from io import StringIO
import io

#可視化ライブラリ
import matplotlib.pyplot as plt
import matplotlib as mp1
import seaborn as sns
sns.set()
%matplotlib inline

#少数第3位まで表記
%precision 3

#ファイルのダウンロード
zip_file_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00356/student.zip'

r = requests.get(zip_file_url, stream=True)
z = zipfile.ZipFile(io.BytesIO(r.content))
z.extractall()

#数学のデータを読み込む
student_data_math = pd.read_csv('student-mat.csv', sep=';')

#groupbyメソッドでreasonを軸にグループ化、sizeメソッドでreasonの値を種類ごとに集計
student_data_math_reason = student_data_math.groupby('reason').size()

#カラー指定
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']

#まとまりからの距離
explode = (0.1, 0, 0, 0) 

# グラフの大きさ指定
plt.figure(figsize = (15, 6))

# グラフを表示
plt.pie(student_data_math_reason.values, explode = explode, labels = student_data_math_reason.index, colors = colors,
        autopct = '%1.1f%%', shadow = True, startangle = -270)

# 円を丸く描画
plt.axis('equal')
```
